```python
%%sql 
SELECT 
    AVG(math_score) AS avg_math , 
    AVG(reading_score) AS avg_reading ,
    AVG(writing_score) AS avg_writing 
FROM 
    studentperformence 
GROUP BY 
    gender ; 
```

```python
%%sql 
SELECT 
    AVG(math_score) AS avg_math , 
    AVG(reading_score) AS avg_reading ,
    AVG(writing_score) AS avg_writing 
FROM 
    studentperformence 
GROUP BY 
    lunch ; 
```

```python
%%sql 
SELECT 
    AVG(math_score) AS avg_math , 
    AVG(reading_score) AS avg_reading ,
    AVG(writing_score) AS avg_writing 
FROM 
    studentperformence 
GROUP BY 
    test ; 
```

```python
%%sql 
SELECT 
    parental_level_of_education , 
    ROUND(AVG((math_score + reading_score + writing_score)/3),2) AS avg_subjects 
FROM 
        studentperformence 
GROUP BY 
    parental_level_of_education 
ORDER BY 
    avg_subjects DESC ;
```

```python
%%sql 
SELECT 
    'math'  as subject , AVG(math_score)  AS avg_score 
FROM 
    studentperformence
UNION 
SELECT 
    'reading',AVG(reading_score) 
FROM 
    studentperformence 
UNION 
SELECT
   'writing',AVG(writing_score) 
FROM 
    studentperformence ;
```

```python
%%sql 
SELECT 
    AVG(math_score) AS avg_math , 
    AVG(reading_score) AS avg_reading ,
    AVG(writing_score) AS avg_writing 
FROM 
    studentperformence 
GROUP BY 
    race_ethnicity;
```

```python
%%sql 
SELECT 
    * , 
    (math_score + reading_score + writing_score ) AS total_score 
FROM 
    studentperformence
ORDER BY 
    total_score DESC ;
```

```python
%%sql 
SELECT 
    gender , 
    SUM( 
        CASE WHEN ( math_score + reading_score + writing_score ) >= 240 THEN 1 ELSE 0 END ) AS high_performance 
FROM 
    studentperformence 
GROUP BY 
    gender ; 
```

```python
%%sql
SELECT 
    gender, 
    SUM(CASE 
            WHEN (math_score + reading_score + writing_score) >= 240 
            THEN 1 
            ELSE 0 
        END) AS high_performance,

    SUM(CASE 
            WHEN (math_score + reading_score + writing_score) >= 180 
             AND (math_score + reading_score + writing_score) < 240
            THEN 1 
            ELSE 0 
        END) AS average_performance,

    SUM(CASE 
            WHEN (math_score + reading_score + writing_score) < 180 
            THEN 1 
            ELSE 0 
        END) AS low_performance

FROM studentperformence
GROUP BY gender;
```
